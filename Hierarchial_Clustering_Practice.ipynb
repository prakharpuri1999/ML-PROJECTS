```python
import numpy as np
X = np.array([[5,3],[10,15],[15,12],[24,10],[30,30],[85,70],[71,80],[60,78],[70,55],[80,91],])
```

```python
import matplotlib.pyplot as plt
labels=range(1,11)
plt.figure(figsize=(10,7))
plt.scatter(X[:,0],X[:,1])
for label,x,y in zip(labels,X[:,0],X[:,1]):
    plt.annotate(label,xy=(x,y),xytext=(-3,3),textcoords='offset points',ha='right',va='bottom')
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
%matplotlib inline
linked=linkage(X,'single')
labelList=range(1,11)
plt.figure(figsize=(10,7))
dendrogram(linked,labels=labelList,orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit_predict(X)
```

```python
cluster.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
import pandas as pd
data=pd.read_csv('shopping-data.csv')
```

```python
data.head()
```

```python
x=data.values[:,3:5]
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
plt.figure(figsize=(10,7))
plt.title('Customer Dendrograms')
dendrogram(linkage(x,method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clf=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
clf.fit_predict(x)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='rainbow')
```
